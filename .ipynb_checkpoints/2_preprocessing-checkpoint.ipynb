# Image Preprocessing Pipeline
This pipeline processes bank check images to enhance readability and extract key zones. Each step plays a crucial role in improving accuracy for OCR and zone detection.

---

# ## Benefits of Each Preprocessing Step

- **Converting to Grayscale**
  - Simplifies processing by removing color information.
  - Enhances contrast for better feature extraction.

- **Applying Gaussian Blur**
  - Reduces image noise and smooths variations.
  - Improves the accuracy of subsequent edge detection.

- **Detecting Edges (Canny Edge Detection)**
  - Highlights important structural components of the image.
  - Helps identify key zones such as text and signature areas.

- **Applying Thresholding**
  - Converts the image to a binary format (black & white).
  - Enhances text readability for OCR and improves zone detection.

- **Morphological Enhancements**
  - Removes small artifacts and noise.
  - Strengthens text regions by dilation and erosion.


```python
import cv2
import numpy as np
import pytesseract
import matplotlib.pyplot as plt
image_path = './data/india/axis bank/en/hand_writing/Cheque 309070.tif'  
image = load_image(image_path)

def show_image(image, title):
    plt.figure(figsize=(8,6))
    plt.imshow(image, cmap='gray')
    plt.title(title)
    plt.axis('off')
    plt.show()

def load_image(image_path):
    image = cv2.imread(image_path)
    show_image(image, "Original Image")
    return image

def convert_to_grayscale(image):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    show_image(gray, "Grayscale Image")
    return gray


def denoise_image(gray):
    blurred = cv2.GaussianBlur(gray, (5,5), 0)
    show_image(blurred, "Denoised Image")
    return blurred

def detect_edges(blurred):
    edges = cv2.Canny(blurred, 50, 150)
    show_image(edges, "Edge Detected Image")
    return edges

def apply_threshold(gray):
    thresh = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)
    show_image(thresh, "Thresholded Image")
    return thresh

def morphological_operations(thresh):
    kernel = np.ones((2,2), np.uint8)
    dilated = cv2.dilate(thresh, kernel, iterations=1)
    eroded = cv2.erode(dilated, kernel, iterations=1)
    show_image(eroded, "Morphological Enhancements")
    return eroded

def process_check(image_path):
    image = load_image(image_path)
    gray = convert_to_grayscale(image)
    blurred = denoise_image(gray)
    edges = detect_edges(blurred)
    thresholded = apply_threshold(gray)
    processed_image = morphological_operations(thresholded)
    return processed_image

image = process_check(image_path)
```
